```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('dados/list_of_countries_by_carbon_dioxide_emissions.csv')
```

```python
dados.head()
```

```python
dados_ = dados.rename(columns = {'Fossil CO2 emissions(Mt CO2)': 'Fossil CO2 emissions(Mt CO2) 1990', 'Unnamed: 2': '2005',
                              'Unnamed: 3': '2017', 'Fossil CO2 emissions': 'Fossil CO2 emissions (2017(% of world))', 
                              'Unnamed: 5': '2017 vs 1990: change (%)', '2017 Fossil CO2 emissions': '2017 Fossil CO2 emissions (Per land area(t CO2/km2/yr))',
                              'Unnamed: 7': 'Per capita(t CO2/cap/yr)', '2018 CO2 emissions': '2018 CO2 emissions (Total including LUCF)',
                              'Unnamed: 9': 'Total excluding LUCF'})
```

```python
dados_.head()
```

```python
dados_.drop(0, axis=0, inplace=True)
```

```python
dados_
```

```python
dados_.info()
```

```python
dadosdf = dados_.drop([1,2,3]).reset_index(drop=True)
dadosdf
```

```python
dadosdf['Fossil CO2 emissions(Mt CO2) 1990'].plot()
```

```python
emissao_com_mais_de_cemvinte_k = dadosdf[dadosdf['Fossil CO2 emissions(Mt CO2) 1990'] >= 120.000]
emissao_com_mais_de_cemvinte_k.head()
```

```python
emissao_com_mais_de_cemvinte_k = emissao_com_mais_de_cemvinte_k.sort_values('Fossil CO2 emissions(Mt CO2) 1990', ascending=False)
```

```python
plt.bar(emissao_com_mais_de_cemvinte_k['Country'], emissao_com_mais_de_cemvinte_k['Fossil CO2 emissions(Mt CO2) 1990'])
plt.xticks(rotation=90)
```
